```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python

# define the mass splittings
mHs = np.arange(60, 140, 10)
mAs = np.arange(60, 500, 10)
```

```python
points = []
for mH in mHs:
    for mA in mAs:
        if ((mH < mA) & (mA - mH >= 20)) & (mA - mH <= 110):
            points.append([mH, mA])
points
```

```python
num_points = 0
for mH in mHs:
    for mA in mAs:
        if ((mH < mA) & (mA - mH >= 20)) & (mA - mH <= 100):
            plt.scatter(mH, mA)
            num_points += 1
plt.xlabel('mH')
plt.ylabel('mA')
#plt.xlim(50, 260)
#plt.ylim(50, 260)
print(f"Final number of points = {num_points}")
```
