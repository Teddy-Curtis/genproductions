```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python

# define the mass splittings
mHs = np.arange(60, 120, 10)
mAs = np.arange(60, 200, 10)
```

```python
num_points = 0
for mH in mHs:
    for mA in mAs:
        if ((mH < mA) & ((mA - mH) <= 80) & (mA - mH > 20)):
            plt.scatter(mH, mA)
            num_points += 1
plt.xlabel('mH')
plt.xlabel('mA')
print(f"Final number of points = {num_points}")
```
